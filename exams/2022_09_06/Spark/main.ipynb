```python
import pyspark
import random
sc = pyspark.SparkContext(appName="Pi")
```

```python
companies_path =  "Companies.txt"
dailyPowerConsumption_path = "DailyPowerConsumption.txt"
dataCenters_path = "DataCenters.txt"
output1 = "solutionOutPart1"
output2 = "solutionOutPart2"
```

```python
# Define the rdds associated with the used input files

# CodDC,CodC,City,Country,Continent
dataCenterRDD = sc.textFile(dataCenters_path).cache()

# CodDC,Date,kWh
pwrConsRDD = sc.textFile(dailyPowerConsumption_path)
```

```python
dataCenterRDD.collect()
```

```python
pwrConsRDD.collect()
```

```python
# Count the total number of data centers worl-wide and compute the threshold (90% of the data centers)
threshold = int(dataCenterRDD.count() * 0.9)
```

```python
highPwrConsDCPerDay = pwrConsRDD.filter(lambda line: float(line.split(",")[2]) >= 1000)\
                    .map(lambda line: (line.split(",")[1], 1))\
                    .reduceByKey(lambda v1, v2: v1 + v2)
```

```python
res1 = highPwrConsDCPerDay.filter(lambda t: t[1] >= threshold)\
                            .keys()
```

```python
res1.saveAsTextFile(output1)
```

```python
# Part2
# Continent(s) with the highest average power consumption per data center in the year
# 2021 and the highest number of data centers. This second part of the application
# selects the continent(s) with (i) the highest average power consumption per data center
# in the year 2021 and (ii) the highest number of data centers. This second part of the
# application selects only the continent(s) satisfying both constraints. Given a continent,
# its average power consumption per data center in 2021 is the sum of the power
# consumption of all its data centers in 2021 divided by the number of data centers in
# that continent. In case of a tie, the application selects all the continents that satisfy both
# constraints. Store the selected continent(s) in the second HDFS output folder (one
# continent per output line).
```

```python
pwrConsRDD.collect()
```

```python
dataCenterRDD.collect()
```

```python
contDC = dataCenterRDD\
    .map(lambda elem: (elem.split(',')[0], elem.split(',')[4]))\
    # .groupByKey()\
    # .mapValues(lambda dc: list(dc))
```

```python
contDC.collect()
```

```python
DCPower = pwrConsRDD.map(lambda elem: (elem.split(',')[0], (float(elem.split(',')[2]), 1)))
```

```python
DCPower.collect()
```

```python
DCContWithCons = contDC.join(DCPower)
```

```python
DCContWithCons.collect()
```

```python
DCPartials = DCContWithCons\
        .map(lambda elem: (elem[1][0], elem[1][1]))\
        .reduceByKey(lambda t1, t2: (t1[0]+t2[0], t1[1]+t2[1]))
```

```python
DCPartials.collect()
```

```python
maxAVG = DCPartials.reduce(lambda el1, el2: max(el1[1][0]/el1[1][1], el2[1][0]/el2[1][1]))
```

```python
maxN = DCPartials.reduce(lambda el1, el2: max(el1[1][1], el2[1][1]))
```

```python
print(maxAVG, maxN)
```

```python
res2 = DCPartials\
    .filter(lambda elem: elem[1][0]/elem[1][1] == maxAVG and elem[1][1] == maxN)\
    .keys()
```

```python
res2.collect()
```
